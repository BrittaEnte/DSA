# 3 How to code the Fibonacci Seequence?

### 3.1 First easy attempt 

```python
def fibonacci(n): 
    if n == 1:
        return 1
    elif n == 2:
        return 1
    elif n > 2:
        return fibonacci(n-1) + fibonacci(n-2)
    
for n in range(1,11):
    print(f' {n} : {fibonacci(n)}')
    
```

### 3.2 Second easy attempt
the problem with the first attemp is that if n gets higher the system needs an tremendous amount of time to calculate. 
this is because every time we call fibonacci n-term, the funtion calls itself until fibonacci is 1.



the solution is to store all the results in a dictionary. so our function just can 
grab the results of all calculations before. 
the process is called MEMORIZATION

```python
cache = {}

def fibonacci(n):
    #check if we got already the value in our cache
    if n in cache:
        return cache[n]
    
    #old function
    if n == 1:
        value = 1
    elif n == 2:
        value = 1
    elif n > 2:
        value = fibonacci(n-1) + fibonacci(n-2)
        
    #cache the value and return it
    cache[n] = value
    return value 
    
for n in range(1,5):
    print(f' {n} : {fibonacci(n)}')

```

### 3.3 LRU Cache
a LRU Cache is Least Recently Used cache
this decorator is imported from the func tools module.
it provides a one-line way to add memorization to your functions


```python
from functools import lru_cache

@lru_cache(maxsize = 1000) # call the dcorator
def fibonacci(n): 
    if n == 1:
        return 1
    elif n == 2:
        return 1
    elif n > 2:
        return fibonacci(n-1) + fibonacci(n-2)
    
for n in range(1,11):
    print(f' {n} : {fibonacci(n)}')
    
```

### 3.4 Make the function rock solid
so far we can enter strings and negative values. 

```python
from functools import lru_cache

@lru_cache(maxsize = 1000) # call the dcorator
def fibonacci(n): 
    #check if the input is a positive integer
    if type(n) != int:
        raise TypeError('n must be an integer')
    if n < 1:
        raise ValueError('n must be positive')
    if n == 1:
        return 1
    elif n == 2:
        return 1
    elif n > 2:
        return fibonacci(n-1) + fibonacci(n-2)
    
for n in range(1,11):
    print(f' {n} : {fibonacci(n)}')   
```
